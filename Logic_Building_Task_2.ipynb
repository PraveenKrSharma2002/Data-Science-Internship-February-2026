#Problem 1: Unique Words in a Sentence

```python
#taking input sentence
sentence = input("Enter sentence: ")

#convert sentence into words
words = sentence.split()

#Remove duplicates using set
unique_words =set(words)
print("Unique words count: ",len(unique_words))
print("Unique words: ", unique_words)
```

#Problem 2: Highest Salary from Employee Data


```python
employees = {"Ravi":75000, "Anita":68000, "Kiran":72000}

#find employee with highest salary
highest_emp = max(employees, key=employees.get)
highest_salary = employees[highest_emp]

print("Highest Salary: ", highest_emp, "_", highest_salary)
```

#Problem 3: Find Maximum and Minimum Values

```python
#Given list
numbers = [45, 22, 89, 10, 66]

#Initialize max and min with first element
max_value = numbers[0]
min_value = numbers[0]

#Loop through list
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

#Print results
print("List:",numbers)
print("Max:",max_value)
print("Min:", min_value)
```

#Problem 4: Count Products Above a Price Threshold

```python
#given list of prices
prices = [450,1200,899,1500,300]

count = 0

#Check each price
for price in prices:
    if price > 1000:
        count +=1

# Print result
print("Products above 1000:",count)
```

#Problem 5: Calculate Attendance Percentage

```python
# Given attendence list
attendence = ["P","P","A","P","P"]

#Count present Days
present_days = attendence.count("P")

#Total Days
total_days = len(attendence)

# Calculate persentage
persentage = (present_days/ total_days)*100

print("Attendence Persentage: ", persentage)
```

#Problem 7: Count Character Frequency

```python
#inout string
text ="ppython"

#create empty dictionary
frequency ={}

# Count each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```
