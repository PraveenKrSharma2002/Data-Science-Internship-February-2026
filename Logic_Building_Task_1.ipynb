#2. Pass Fail Analyzer ðŸ‘‡

```python
#Given marks list
marks = [45,78,90,33,60]

pass_count =0
fail_count =0

#checking each student Mark
for mark in marks:
  if mark >=50:
    pass_count +=1
  else:
    fail_count +=1

#Printing the Final Result
print("Total Student : ",len(marks))
print("Total Pass Students :",pass_count)
print("Total Fail StudentS :",fail_count)
```

#3. Simple Data Cleaner ðŸ‘‡

```python
#Given List
names = [" Alice ","bob"," CHARLIE "]

cleaned_names =[]

for name in names:
  cleaned_name = name.strip().lower()  # Remove spaces + convert to lower case
  cleaned_names.append(cleaned_name)

print("Cleaned Names List : ", cleaned_names)
```

#4. Message Length Analyzer ðŸ‘‡

```python
#Given message list
messages =["Hi","Welcome to the platform","OK"]

for message in messages :
  length =len(message)
  print("Message : ", message)
  print("Length :",length)

  if length > 10:
    print("Status : Long message")
  else:
    print("Status : Short message")

  print("============================================")
```
